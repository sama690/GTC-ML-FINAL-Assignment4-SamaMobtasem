```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=["median_house_value"])

# Target (the column you want to predict)
y = df["median_house_value"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2)
```

```python
#after seeing the histogram we recognized that some of the graphs are shifted to the left so we do a process similar to normalization but we use the logarithmic function
```

```python
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
```

```python
#third: A relaition between where the house hold is located and the median value for it (target)
#using melt
```

```python
melted_df = pd.melt(
    train_data,
    id_vars=['median_house_value'],
    value_vars=['<1H OCEAN', 'INLAND', 'ISLAND', 'NEAR BAY', 'NEAR OCEAN'],
    var_name='ocean_category',
    value_name='is_present'
)
filtered_df = melted_df[melted_df['is_present'] == 1]
plt.figure(figsize=(12, 6))
sns.boxplot(x='median_house_value',y='ocean_category', data=filtered_df)
plt.title('Median House Value by Ocean Proximity Category')
plt.xlabel('Ocean Proximity Category')
plt.ylabel('Median House Value')
plt.show()
```

```python
train_data['bedrooms_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms'] / train_data['households']
```

```python
train_data.dropna(inplace=True)
```

```python
#Training and testing in Linear Regresstion
# all have (train=80% and test=20%)
```

```python
#Linear Regresstion
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
reg= LinearRegression()

X = train_data.drop(columns=["median_house_value"])
y = train_data["median_house_value"]

X_train, X_test, y_train, y_test = train_test_split(
   X, y, test_size=0.2)
reg.fit(X_train,y_train)
```

```python
#preparing the test data
```

```python
test_data = X_test.join(y_test)
test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms_ratio'] = test_data['total_rooms'] / test_data['households']
X = test_data.drop('median_house_value', axis=1)
y = test_data['median_house_value']
```

```python
#the Linear Regresstion is accurte with a 66.11 %
```

```python
reg.score(X_test,y_test)
```

```python
#Random forest has a better score as it is built using multiple desition trues
```

```python
#Random Forest
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
#Improving using Cross Validation and Grid Search
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'min_samples_split': [2, 4],
    'max_depth': [None, 4, 8]
}

# Fix: The param_grid argument should be explicitly named
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,  # Corrected line
    cv=5,
    scoring='neg_mean_squared_error',
    return_train_score=True
)

# This line is correct, assuming X_train and y_train are defined
grid_search.fit(X_train, y_train)
```

```python
#Random forest was found to be better
```

```python
grid_search.best_estimator_
```
